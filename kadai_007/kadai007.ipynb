```python
# リスト（array）
array = [
    "月曜日は晴れです",
    "火曜日は雨です",
    "水曜日は晴れです",
    "木曜日は晴れです",
    "金曜日は曇りです",
    "土曜日は曇りのち雨です",
    "日曜日は雷雨です"
]

# ディクショナリ（dictionary） ※課題の指定通り：変数名は dictionary、キーは英語省略
dictionary = {
    "mon": "月曜日は晴れです",
    "tue": "火曜日は雨です",
    "wed": "水曜日は晴れです",
    "thu": "木曜日は晴れです",
    "fri": "金曜日は曇りです",
    "sat": "土曜日は曇りのち雨です",
    "sun": "日曜日は雷雨です"
}

# 水曜日の天気予報を取り出して出力（リスト）
print(array[2])

# 水曜日の天気予報を取り出して出力（ディクショナリ）
print(dictionary["wed"])
```
