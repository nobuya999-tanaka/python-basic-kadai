```python
# リスト（array）：順番で管理、全文を保持
array = [
    "月曜日は晴れです",
    "火曜日は雨です",
    "水曜日は晴れです",
    "木曜日は晴れです",
    "金曜日は曇りです",
    "土曜日は曇りのち雨です",
    "日曜日は雷雨です"
]

# ディクショナリ（dictionary）：キーで管理、天気のみを保持
dictionary = {
    "mon": "晴れ",
    "tue": "雨",
    "wed": "晴れ",
    "thu": "晴れ",
    "fri": "曇り",
    "sat": "曇りのち雨",
    "sun": "雷雨"
}

# 水曜日の天気予報を取り出して出力
print(array[2])                  # リストから全文
print("水曜日は" + dictionary["wed"] + "です")  # ディクショナリから天気のみ
```
