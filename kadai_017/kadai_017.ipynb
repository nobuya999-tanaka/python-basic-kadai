```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は大人ではありません。")


# Humanクラスのインスタンスを複数生成
humans = [
    Human("伸弥", 25),
    Human("萌楓", 18),
    Human("清美", 20),
    Human("樹", 15)
]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
