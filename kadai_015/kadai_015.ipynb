```python
# Humanクラスの定義
class Human:
    # コンストラクタ（初期化メソッド）
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # name と age を標準出力するメソッド
    def printinfo(self):
        print("名前:", self.name)
        print("年齢:", self.age)


# Humanクラスのインスタンスを変数に代入
human1 = Human("田中伸弥", 35)

# メソッドを呼び出して情報を表示
human1.printinfo()
```
